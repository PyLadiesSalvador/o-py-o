## CRIAR NOVA COLUNA COM VALOR CALCULADO DO IMC

```python
df.head(1)
```

```python
# o imc é calculado dividindo o peso (em kg) pela altura ao quadrado (em metros).
df['imc'] = (df['peso']/(df['altura']**2)).round(2)
df
```

## CLASSIFICAR DADOS (SE ... ENTÃO)

Classificação IMC:
IMC < 18,5 - abaixo do peso normal
18,5 <= IMC < 25 - normal
25 <= IMC < 30 - sobrepeso
30 <= IMC < 35 - obesidade
35 <= IMC < 40 - obesidade severa
IMC > 40 - obseidade mórbida

```python
# classificação imc:
df.head(1)
```

```python
condições = [
    df['imc'] < 18.5,
    df['imc'] < 25,
    df['imc'] < 30,
    df['imc'] < 35,
    df['imc'] < 40,
    df['imc'] >= 40
]

escolhas = [
    'abaixo do peso normal',
    'normal',
    'sobrepeso',
    'obesidade',
    'obesidade severa',
    'obesidade mórbida'
]

df['classificação'] = np.select(condições, escolhas, default = 'olhar')
df
```

```python
df.head(1)
```

Classificação IMC Feminino:
IMC < 19 - abaixo do peso normal
19 <= IMC < 24 - normal
24 <= IMC < 29 - sobrepeso
29 <= IMC < 34 - obesidade
34 <= IMC < 39 - obesidade severa
IMC >= 39 - obseidade mórbida

Classificação IMC Masculino:
IMC < 20 - abaixo do peso normal
20 <= IMC < 25 - normal
25 <= IMC < 30 - sobrepeso
30 <= IMC < 35 - obesidade
35 <= IMC < 40 - obesidade severa
IMC >= 40 - obseidade mórbida

```python
condições = [
    (df['imc'] < 19) & (df['sexo'] == 'feminino'),
    (df['imc'] < 24) & (df['sexo'] == 'feminino'),
    (df['imc'] < 29) & (df['sexo'] == 'feminino'),
    (df['imc'] < 34) & (df['sexo'] == 'feminino'),
    (df['imc'] < 39) & (df['sexo'] == 'feminino'),
    (df['imc'] >= 39) & (df['sexo'] == 'feminino'),
    (df['imc'] < 20) & (df['sexo'] == 'masculino'),
    (df['imc'] < 25) & (df['sexo'] == 'masculino'),
    (df['imc'] < 30) & (df['sexo'] == 'masculino'),
    (df['imc'] < 35) & (df['sexo'] == 'masculino'),
    (df['imc'] < 40) & (df['sexo'] == 'masculino'),
    (df['imc'] >= 40) & (df['sexo'] == 'masculino')
    ]

escolhas = [
    'abaixo do peso normal',
    'normal',
    'sobrepeso',
    'obesidade',
    'obesidade severa',
    'obesidade mórbida',
    'abaixo do peso normal',
    'normal',
    'sobrepeso',
    'obesidade',
    'obesidade severa',
    'obesidade mórbida'
]

df['classificação_fem_mas'] = np.select(condições, escolhas, default = 'olhar')
df
```

```python
df['classificação'] = df['classificação'].astype("category")
```
